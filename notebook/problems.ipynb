## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
# Define all the probability functions related to this distribution

def dnorm(x, mean=0, std=1):
    """
    Calcula la densidad de la distribución normal en el punto dado.

    Parámetros:
    x (float o array-like): Punto o puntos donde calcular la densidad.
    mean (float): Media de la distribución normal.
    std (float): Desviación estándar de la distribución normal.

    Devuelve:
    float o array-like: Densidad de la distribución normal en el punto o puntos dados.
    """
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def pnorm(x, mean=0, std=1):
    """
    Calcula la función de distribución acumulativa (CDF) de la distribución normal.

    Parámetros:
    x (float o array-like): Punto o puntos donde calcular la CDF.
    mean (float): Media de la distribución normal.
    std (float): Desviación estándar de la distribución normal.

    Devuelve:
    float o array-like: CDF de la distribución normal en el punto o puntos dados.
    """
    result = norm.cdf(x=x, loc=mean, scale=std)
    return result

def qnorm(p, mean=0, std=1):
    """
    Calcula la función cuantil (o inversa) de la distribución normal.

    Parámetros:
    p (float o array-like): Probabilidad o probabilidades en las que calcular el cuantil.
    mean (float): Media de la distribución normal.
    std (float): Desviación estándar de la distribución normal.

    Devuelve:
    float o array-like: Cuantil de la distribución normal para la probabilidad dada.
    """
    result = norm.ppf(q=p, loc=mean, scale=std)
    return result

def rnorm(n, mean=0, std=1):
    """
    Genera muestras aleatorias de la distribución normal.

    Parámetros:
    n (int): Número de muestras a generar.
    mean (float): Media de la distribución normal.
    std (float): Desviación estándar de la distribución normal.

    Devuelve:
    array-like: Muestras aleatorias de la distribución normal.
    """
    result = norm.rvs(size=n, loc=mean, scale=std)
    return result
    
np.random.seed(42)


# Densidad de la distribución normal
print(f"Densidad de la distribución normal en x = 2: {dnorm(2, mean=10*0.2, std=np.sqrt(10*0.2*(1-0.2)))}")

# Función de distribución acumulativa (CDF) de la distribución normal
print(f"Probabilidad de que una distribución normal sea menor que x = 2: {pnorm(2, mean=10*0.2, std=np.sqrt(10*0.2*(1-0.2)))}")
# Cuantil de la distribución normal
print(f"Cuál es el valor de la distribución normal con probabilidad acumulada de 0.9?: {qnorm(0.9, mean=10*0.2, std=np.sqrt(10*0.2*(1-0.2)))}")
# Generar valores aleatorios de la distribución normal
print(f"Generar 2000 valores aleatorios de la distribución normal: {rnorm(2000, mean=10*0.2, std=np.sqrt(10*0.2*(1-0.2)))}")


```
